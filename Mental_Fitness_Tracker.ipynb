## Transforming non-numeric labels to numeric labels

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
for i in data.columns:
    if data[i].dtype=='object':
        data[i]=l.fit_transform(data[i])
```

```python
data.shape
```

## Splitting the data into train and test data

```python
x=data.drop('mental_fitness',axis=1)
y=data['mental_fitness']

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print('x_train : ',x_train.shape)
print('x_test : ',x_test.shape)
print('\n y_train : ',y_train.shape)
print('y_test : ',y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_train_pred=lr.predict(x_train)
mse=mean_squared_error(y_train,y_train_pred)
rmse=(np.sqrt(mean_squared_error(y_train,y_train_pred)))
r2=r2_score(y_train,y_train_pred)
```

```python
print('The Linear Regressin model performance for training set ')
print('----------------------------------------------------------')
print('MSE is {}'.format(mse))
print('RSME is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
```

```python
y_train_pred=rf.predict(x_train)
mse=mean_squared_error(y_train,y_train_pred)
rmse=(np.sqrt(mean_squared_error(y_train,y_train_pred)))
r2=r2_score(y_train,y_train_pred)
```

```python
print('The Random Forest Regressor model performance for training set')
print('---------------------------------------------------')
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

## Evaluation

## Linear Regression model evaluation for testing data

```python
y_test_pred=lr.predict(x_test)
mse=mean_squared_error(y_test,y_test_pred)
rmse=(np.sqrt(mean_squared_error(y_test,y_test_pred)))
r2=r2_score(y_test,y_test_pred)
```

```python
print('The Linear Regressin model performance for testing set')
print('-----------------------------------------------------')
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```

## Random forest regressor model evaluation for testing set

```python
y_test_pred=rf.predict(x_test)
mse=mean_squared_error(y_test,y_test_pred)
rmse=(np.sqrt(mean_squared_error(y_test,y_test_pred)))
r2=r2_score(y_test,y_test_pred)
```

```python
print('The Random Forest Regressr model performance for testing set')
print('------------------------------------------------------------')
print('MSE is {}'.format(mse))
print('RMSE is {}'.format(rmse))
print('R2 score is {}'.format(r2))
```
